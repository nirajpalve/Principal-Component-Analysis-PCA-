```python
import pandas as pd  # Importing Libraries 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
Wine = pd.read_csv('wine.csv')    # Loading Dataset
Wine 
```

```python
Wine['Type'].value_counts()
```

```python
Wine.info()  # To check wheather null value exist or not.
```

```python
Wine_1 = Wine.iloc[:,1:]  # Removing Column 1
Wine_1
```

```python
Wine_array = Wine_1.values    # Converting data to numpy array
Wine_array
```

```python
Wine_norm = scale(Wine_array)    # Normalizing data 
Wine_norm
```

# PCA 

```python
pca = PCA(n_components = 13)  # Applying PCA Fit Transform to dataset
Wine_pca = pca.fit_transform(Wine_norm)
Wine_pca
```

```python
pca.components_   # PCA compenent matrix
```

```python
Var = pca.explained_variance_ratio_   # The amount of variance that each PCA has
Var
```

```python
Var_1 = np.cumsum(np.round(Var,4)*100)    # Cummulative variance of each PCA
Var_1
```

```python
plt.plot(Var_1,color='green')  #  Variance plot for PCA components obtained
plt.show()   
```

```python
# Final Dataframe
final_df = pd.concat([Wine['Type'], pd.DataFrame(Wine_pca[:,0:3], columns = ['PC1','PC2','PC3'])], axis = 1 )
final_df
```

```python
fig = plt.figure(figsize = (10,6))
sns.scatterplot(data = final_df)
```

# Checking with other Clustering Algorithms

# Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch        # Import Libraries
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,8))
Dendrogram = sch.dendrogram(sch.linkage(Wine_norm, 'average'))
```

```python
hclusters = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'average')  # Create Clusters
hclusters
```

```python
y = pd.DataFrame(hclusters.fit_predict(Wine_norm), columns = ['clustersid'])
y['clustersid'].value_counts()
```

```python
Wine_2 = Wine.copy()      # Adding clusters to dataset
Wine_2['clustersid'] = hclusters.labels_
Wine_2
```

# K-Means Clustering

```python
from sklearn.cluster import KMeans    # Import Libraries
```

```python
wcss = []                              # Within-cluster sum-of-squares criterion 
for i in range (1,6):
    kmeans = KMeans(n_clusters = i,random_state = 2)
    kmeans.fit(Wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

# Build Cluster algorithm using K=3

```python
Clusters_1 = KMeans(3,random_state = 30).fit(Wine_norm)   # Cluster algorithm using K=3
Clusters_1
```

```python
Clusters_1.labels_
```

```python
Wine_2 = Wine.copy()           # Assign clusters to the data set
Wine_2['clusters3id']=Clusters_1.labels_
Wine_2
```
